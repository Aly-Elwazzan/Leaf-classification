```
import numpy as np
import pandas as pd
import itertools
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.linear_model import SGDClassifier
```

```
# Get train and test data
dftest=pd.read_csv('/gdrive/My Drive/leaf/test.csv')
dftrain=pd.read_csv('/gdrive/My Drive/leaf/train.csv')
```

```
train_id=dftrain.id
train_id.head()
```

```
#getting labels of  train
train_labels=dftrain.species
train_labels.head()
```

```
train_features=dftrain
train_features=train_features.drop(['species', 'id'], axis=1) 
train_features.head()
```

```
#splitting but not stratified
x_train,x_test,y_train,y_test=train_test_split(train_features, train_labels, test_size=0.2, random_state=7)
```

```
from sklearn.linear_model import SGDClassifier
#Initialize stochastic Gradient Decent
sgd_clf = SGDClassifier(max_iter=5, tol=-np.infty, random_state=42)
sgd_clf.fit(x_train, y_train)
```

```
# Predict on the test set and calculate accuracy
y_pred=sgd_clf.predict(x_test)
score=accuracy_score(y_test,y_pred)
print(f'Accuracy: {round(score*100,2)}%')
```

```
#split  SSS
from sklearn.model_selection import StratifiedShuffleSplit
sss = StratifiedShuffleSplit(n_splits=10, test_size=0.2, random_state=23)
#sss.get_n_splits(train_features, labels)

for train_index, test_index in sss.split(train_features,train_labels ):
   X_train, X_test = train_features.values[train_index], train_features.values[test_index]
   Y_train, Y_test = train_labels[train_index], train_labels[test_index]
```

```
from sklearn.linear_model import SGDClassifier
#Initialize stochastic Gradient Decent
sgd_clf = SGDClassifier(max_iter=50, penalty=None,eta0=0.1, random_state=42)
sgd_clf.fit(X_train, Y_train)
```

```
Y_pred=sgd_clf.predict(X_test)
score=accuracy_score(Y_test,Y_pred)
print(f'Accuracy: {round(score*100,2)}%')
```
